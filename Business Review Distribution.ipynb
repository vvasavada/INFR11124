```python
import networkx as nx
import matplotlib.pyplot as ply

%matplotlib inline
```

```python
import parser
```

```python
businesses = parser.getBusinesses()
review_counts = []
for business in businesses:
    review_counts.append(business.getReviewCount())
    
nonzero_review_counts = [n for n in review_counts if n > 0]
```

```python
from collections import Counter

frequency = dict(Counter(nonzero_review_counts))

fig = ply.figure()
ax = ply.gca()
fig.suptitle('Distribution of reviews')
ax.scatter(frequency.keys(), [frequency[n] for n in frequency.keys()])
ply.xlabel("Reviews")
ply.ylabel("Number of businesses")
ply.show()

fig = ply.figure()
ax = ply.gca()
fig.suptitle('Log-Log Scale: Distribution of average reviews per business')
ax.scatter(frequency.keys(), [frequency[n] for n in frequency.keys()])
ply.xlabel("Reviews")
ply.ylabel("Number of businesses")
ax.set_yscale('log')
ax.set_xscale('log')
```
