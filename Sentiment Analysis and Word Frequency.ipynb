```python
import networkx as nx
import matplotlib.pyplot as ply

%matplotlib inline
```

```python
fig = ply.figure()
ax = ply.gca()

fig.suptitle('Relation between Word Frequency and Review Votes')
ax.scatter(frequency_votes.keys(), [frequency_votes[n] for n in frequency_votes.keys()])
ply.xlabel("Frequency")
ply.ylabel("Votes")
ply.show()
```
