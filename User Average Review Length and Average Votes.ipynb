```python
import networkx as nx
import matplotlib.pyplot as ply

%matplotlib inline
```

```python
from extract_user_reviews import *
```

```python
import parser
```

```python
users = parser.getUsers()
avg_vote_counts = {}
for user in users:
        avg_vote_counts[user.getID()] = user.getAverageVotes()
        
lengths = []
votes = []

for user in user_average_review_length:
    lengths.append(user_average_review_length[user])
    votes.append(avg_vote_counts[user])
```

```python
fig = ply.figure()
ax = ply.gca()
fig.suptitle('Average votes and average review lengths per user')
ax.scatter(lengths, votes)
ply.xlabel("Average Lengths")
ply.ylabel("Average Votes")
ply.show()
```
